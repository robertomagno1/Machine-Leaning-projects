```python
#cluster analysis::

from sklearn.cluster import KMeans
import numpy as np

# Preparazione dei dati per l'analisi per cluster
# Calcolo della media della percentuale di superficie biologica per paese
bio_area_by_country = filtered_bio_area.groupby('geo')['Bio_Area'].mean().reset_index()

# Utilizzo del metodo del gomito per determinare il numero ottimale di cluster
sse = {}
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42).fit(bio_area_by_country[['Bio_Area']])
    sse[k] = kmeans.inertia_  # somma degli errori quadratici

# Plot del metodo del gomito
plt.figure(figsize=(10, 6))
plt.plot(list(sse.keys()), list(sse.values()), 'bx-')
plt.xlabel("Numero di cluster")
plt.ylabel("SSE (somma degli errori quadratici)")
plt.title("Metodo del Gomito per Determinare il Numero Ottimale di Cluster")
plt.show()

# Procediamo con l'analisi dei cluster basata su 3 cluster (numero comunemente scelto)
kmeans_final = KMeans(n_clusters=3, random_state=42)
bio_area_by_country['Cluster'] = kmeans_final.fit_predict(bio_area_by_country[['Bio_Area']])

# Visualizzazione dei cluster
plt.figure(figsize=(12, 6))
sns.scatterplot(data=bio_area_by_country, x='geo', y='Bio_Area', hue='Cluster', palette='viridis', s=100)
plt.title('Cluster di Paesi basati sulla Percentuale di Superficie Agricola Biologica')
plt.xlabel('Paese')
plt.ylabel('Media Percentuale di Superficie Agricola Biologica')
plt.xticks(rotation=90)
plt.legend(title='Cluster')
plt.show()

bio_area_by_country.sort_values(by='Cluster')  # Visualizzazione dei dati di cluster ordinati
```

#Ho eseguito un'analisi per cluster per raggruppare i paesi in base alla percentuale media di superficie agricola biologica utilizzando l'algoritmo K-means. Dall'analisi del metodo del gomito, ho scelto di utilizzare 3 cluster per questa analisi, poiché sembra offrire un buon equilibrio.



Risultati dei Cluster:

Cluster 1: Contiene paesi con percentuali molto basse di superficie agricola biologica (generalmente sotto il 5%). Questo cluster include paesi come Albania (AL), Islanda (IS), e Regno Unito (UK).
Cluster 2: Include paesi con percentuali medie di superficie agricola biologica (dal 5% al 10%). Paesi come Germania (DE), Spagna (ES), e Danimarca (DK) sono in questo gruppo.
Visualizzazione dei Cluster:
La mappa di dispersione mostra i paesi colorati in base al cluster di appartenenza. Questo aiuta a visualizzare come i diversi paesi si raggruppano in base alla loro percentuale di superficie biologica.



Questi cluster possono aiutare a identificare e analizzare le politiche agricole, i supporti economici e le condizioni ambientali che influenzano il livello di adozione dell'agricoltura biologica. Ogni cluster potrebbe meritare un'indagine specifica per comprendere le forze motrici dietro le differenze osservate.

```python
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import seaborn as sns

# Assumendo che 'data_ricerca_clean' e 'data_bio_clean' siano già definiti e puliti, e che includano le colonne necessarie

# Unione dei DataFrame
combined_filtered_data = pd.merge(data_ricerca_clean, data_bio_clean, on=['geo', 'TIME_PERIOD'], how='inner')

# Verifica e pulizia dei dati mancanti
combined_filtered_data.dropna(subset=['Investment', 'Bio_Area'], inplace=True)

# Standardizzazione dei dati
scaler = StandardScaler()
data_scaled = scaler.fit_transform(combined_filtered_data[['Investment', 'Bio_Area']])

# Applicazione di PCA
pca = PCA(n_components=2)
principal_components = pca.fit_transform(data_scaled)

# Visualizzazione dei risultati
plt.figure(figsize=(10, 6))
sns.scatterplot(x=principal_components[:, 0], y=principal_components[:, 1])
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA of Investment and Bio Area')
plt.grid(True)
plt.show()

# Stampare la varianza spiegata
print("Explained Variance:", pca.explained_variance_ratio_)
```

Interpretazione del Grafico PCA
Il grafico risultante dalla PCA che visualizza le due componenti principali avrà i seguenti aspetti chiave da considerare:

Assi:
Asse X (Componente Principale 1): Rappresenta la direzione di massima varianza nei dati. La maggior parte delle informazioni (varianza) dei dati originali è catturata lungo questo asse.
Asse Y (Componente Principale 2): Rappresenta la seconda direzione più significativa di varianza, ortogonale alla prima componente.
Punti:
Ogni punto nel grafico rappresenta un'osservazione (in questo caso, un'aggregazione di dati per paese e anno) proiettata nelle due principali dimensioni ridotte.
La distanza tra i punti può indicare la somiglianza o la differenza nelle caratteristiche di investimento e superficie biologica tra i paesi/anni.
Distribuzione e Raggruppamento:
Se i punti sono raggruppati insieme, ciò indica che le osservazioni hanno caratteristiche simili in termini di investimenti e superficie biologica.
Se i punti sono sparsi ampiamente, indica una grande varietà nelle caratteristiche tra i paesi/anni.
Varianza Spiegata:
La varianza spiegata da ciascuna componente principale, che hai stampato come output, ti dirà quanto della varianza totale dei dati originali è catturata da queste due componenti. Ad esempio, se il primo valore è 0.70, significa che il 70% della varianza totale è catturato dalla prima componente.
Domande da Porsi:
Quanto bene le due componenti principali rappresentano i dati originali? Se la somma delle varianze spiegate è alta (ad esempio > 80%), le due componenti danno una buona rappresentazione.
Ci sono raggruppamenti chiari o pattern nel grafico? Questo potrebbe indicare relazioni o tendenze significative tra gli investimenti e l'estensione dell'agricoltura biologica.
Quali sono le implicazioni per le politiche o le decisioni future? Ad esempio, se alcuni paesi con investimenti simili hanno superfici biologiche diverse, ciò potrebbe indicare l'efficacia relativa delle politiche.

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Preparazione dei dati per il modello di regressione
# Assumendo che 'combined_filtered_data' contenga le colonne 'Investment' e 'Bio_Area' pulite e pronte per l'uso
X = combined_filtered_data[['Investment']]  # Variabile predittiva
y = combined_filtered_data['Bio_Area']  # Variabile risposta

# Suddividere i dati in set di training e test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Costruzione del modello di regressione lineare
model = LinearRegression()
model.fit(X_train, y_train)

# Predizione sui dati di test
y_pred = model.predict(X_test)

# Calcolo delle metriche di performance
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

# Output delle metriche
mse, r2
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#procedo con analisi espolartiva dei dati EDA

# Rimuovi i dati aggregati dell'Unione Europea ("EU27_2020") per evitare distorsioni nell'analisi
data_no_eu_agg = data_aggregated[data_aggregated['geo'] != 'EU27_2020']

# Calcolo della media degli investimenti e dell'area biologica per ogni anno come rappresentativo per l'UE
eu_mean = data_no_eu_agg.groupby('TIME_PERIOD').agg({
    'Investment': 'mean',
    'Bio_Area': 'mean'
}).reset_index()
eu_mean['geo'] = 'EU_Mean'

# Aggiungi la media calcolata al dataset escludendo il valore complessivo
data_corrected = pd.concat([data_no_eu_agg, eu_mean], ignore_index=True)

# Scatter plot senza EU27_2020 e con la media UE inclusa
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data_corrected, x='Investment', y='Bio_Area', hue='geo', palette='viridis', s=100)
plt.title('Investimenti vs Area Biologica (senza EU27_2020, con Media UE)')
plt.xlabel('Investimenti in Ricerca e Sviluppo (€ per abitante)')
plt.ylabel('Percentuale di Superficie Agricola Biologica')
plt.legend(title='Paese', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()

# Line plot per paese senza EU27_2020 e con la media UE
fig, axes = plt.subplots(nrows=6, ncols=6, figsize=(18, 10), sharex=True, sharey=True)
axes = axes.flatten()
countries = data_corrected['geo'].unique()

for i, country in enumerate(countries):
    if i >= len(axes):  # Verifica di non superare il numero di subplot disponibili
        break
    subset = data_corrected[data_corrected['geo'] == country]
    sns.lineplot(ax=axes[i], data=subset, x='TIME_PERIOD', y='Investment', color='blue', marker='o', label='Investment')
    sns.lineplot(ax=axes[i], data=subset, x='TIME_PERIOD', y='Bio_Area', color='green', marker='o', label='Bio Area')
    axes[i].set_title(country)
    axes[i].legend(loc='upper left')

plt.tight_layout()
plt.show()
```
